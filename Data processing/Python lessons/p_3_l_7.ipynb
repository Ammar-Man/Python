```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# obs enbart 1000 räcker inte för att lagra ljud!
t = np.linspace(0,6.3, num=4000)
# t[2000:2300]
```

```python
# Sinus rörö sig mellan
toneA = np.sin(t)
toneA[2100:2103]
```

```python
plt.plot(t,toneA)
```

```python
# enhetscirkeln med sinus och cosinus 
toneB = np.cos(t)
```

```python
# sinus som x och cosinun som y 
plt.plot(toneA, toneB)
```

```python
# toneA.tofile?
# a.tofile(fid, sep="", format="%s")
```

```python
# expotera file till en "excel fil"
toneA.tofile("toneA.csv", sep=",")
```

```python
# skriv ut en wavfil med hjälp scipy 
from scipy.io.wavfile import write
```

```python
# write?
# write(filename, rate, data) 
```
