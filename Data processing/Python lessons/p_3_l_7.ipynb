```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from scipy.io.wavfile import write
```

```python
# C4 = 261.63
# G4 = 392.00 hz
# E4 = 329.63 hz

time = np.linspace(0,1, num=16000)

# print(len(C4_t))
# print(C4_t[0:10])
# t[2000:2300]
#  1s = 261.63 perisoden av sinus 
# Sinus rr sig mellan
# t[2000:2300]
tone_C4t = np.sin(2 * np.pi * 261 * time )
tone_G4t = np.sin(2 * np.pi * 392 * time )
tone_E4t = np.sin(2 * np.pi * 329 * time )
# tone_C4t[2000:2200]
print(len(tone_C4t))
# plt.plot(t)
# plt.plot(toneA)

plt.plot(time, tone_C4t)
plt.plot(time, tone_G4t)
plt.plot(time, tone_E4t)
plt.xlim(0,0.02)

C_accord = tone_C4t + tone_G4t + tone_E4t 

# plt.plot(C4_t,E4_t)

#  C-accord = CEG
```

```python
C_accord = tone_C4t + tone_G4t + tone_E4t
plt.plot(time, C_accord )


# skriv ut en wavfil med hjlp scipy 
from scipy.io.wavfile import write
write("C_accord.wav",16000 , C_accord)
```

```python
# obs enbart 1000 rcker inte fr att lagra ljud!
t = np.linspace(0,500, num=65000)
print(len(t))
# t[2000:2300]

# Sinus rr sig mellan
# t[2000:2300]
toneA = np.sin(t)
# toneA[100:2000:-1]
print(len(toneA))
# plt.plot(t)
# plt.plot(toneA)
toneA.tofile("toneCtest.csv", sep=",")

plt.plot(t,toneA)

```
